```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df_detractor = pd.read_excel(r'C:\Users\tuteg\Desktop\Strata Analytics - Challenge\avg-services.xlsx', sheet_name='Detractor')
df_passive = pd.read_excel(r'C:\Users\tuteg\Desktop\Strata Analytics - Challenge\avg-services.xlsx')
df_promoter = pd.read_excel(r'C:\Users\tuteg\Desktop\Strata Analytics - Challenge\avg-services.xlsx', sheet_name='Promoter')
```

```python
df_detractor = df_detractor.sort_values('Score', ascending=True).head(5)
df_passive = df_passive.sort_values('Score', ascending=True).head(5)
df_promoter = df_promoter.sort_values('Score', ascending=True).head(5)

df_promoter.head()
```

```python
# Detractor
X_detractor = list(df_detractor.iloc[:, 0])
Y_detractor = list(df_detractor.iloc[:, 1])

# Passive
X_passive = list(df_passive.iloc[:, 0])
Y_passive = list(df_passive.iloc[:, 1])

# Promoter
X_promoter = list(df_promoter.iloc[:, 0])
Y_promoter = list(df_promoter.iloc[:, 1])
```

```python
# Detractor: Plot the data using bar() method
plt.bar(X_detractor, Y_detractor, color='r')
plt.title("Detractor Score Services")
plt.xticks(ticks=range(len(df_detractor)), labels=X_detractor, rotation=90)
plt.xlabel("Services")
plt.ylabel("Score")
```

```python
# Passive: Plot the data using bar() method
plt.bar(X_passive, Y_passive, color='b')
plt.title("Passive Score Services")
plt.xticks(ticks=range(len(df_passive)), labels=X_passive, rotation=90)
plt.xlabel("Services")
plt.ylabel("Score")
```

```python
# Promoter: Plot the data using bar() method
plt.bar(X_promoter, Y_promoter, color='g')
plt.title("Promoter Score Services")
plt.xticks(ticks=range(len(df_promoter)), labels=X_promoter, rotation=90)
plt.xlabel("Services")
plt.ylabel("Score")
```
